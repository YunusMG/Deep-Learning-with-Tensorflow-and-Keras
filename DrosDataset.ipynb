```python
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2


DATADIR = "F:/DeepLearningPython/model_organisms"
CATEGORIES = ['Drosophila_dataset', 'zebrafish_dataset']


for category in CATEGORIES:
    path = os.path.join(DATADIR,category) #paths to Dros and fish dir
    for img in os.listdir(path):
        img_array = cv2.imread(os.path.join(path,img), cv2.IMREAD_GRAYSCALE)
        plt.imshow(img_array, cmap="gray")
        plt.show()
        
        break
        
    break
```

```python
IMG_SIZE = 70
new_array = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE))
plt.imshow(new_array, cmap = 'gray')
plt.show()
```

```python
training_data = []

def create_training_data():
    for category in CATEGORIES:
        path = os.path.join(DATADIR, category) #paths to Dros and fish dir
        class_num = CATEGORIES.index(category)
        for img in os.listdir(path):
            try:
                img_array = cv2.imread(os.path.join(path,img), cv2.IMREAD_GRAYSCALE)
                new_array = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE))
                training_data.append([new_array, class_num])
            except Exception as e:
                pass
create_training_data()
```

```python
print(len(training_data))
```

```python
import random
random.shuffle(training_data)
```

```python
for sample in training_data:
    print(sample[1])
```

```python
D = []
f = []
```

```python
for features, label in training_data:
    D.append(features)
    f.append(label)
    
D = np.array(D).reshape(-1, IMG_SIZE, IMG_SIZE, 1)
```

```python
import pickle

pickle_out = open("D.pickle","wb")
pickle.dump(D, pickle_out)
pickle_out.close()

pickle_out = open("f.pickle","wb")
pickle.dump(f, pickle_out)
pickle_out.close()
```
